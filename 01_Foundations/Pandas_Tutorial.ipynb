```python
## Series
## Syntax
## varname= pandas.Series(object,index,dtype)

a=[100,200,300,400]
print(pd.Series(a,index=[i for i in range(1,len(a)+1)],dtype=float))
print(type(a))

s="INDIA"
print(type(s))
print(pd.Series(list(s)))

s1={100,200,300,400,500}
#print(pd.Series(s1)) # TypeError: 'set' type is unordered

print(pd.Series(tuple(s1)))
print(pd.Series(list(s1)))
print(pd.Series([s1]))
```

```python
## Create  a Series using DICT

s={10: "Python",20:"Django",30:"AI",40:"ML",50:"DL"}
print(pd.Series(s))
```

```python
import numpy as np
vec1= np.asarray([2,3,4])
vec2=np.asarray([3,4,5])

dot_np= np.dot(vec1,vec2)
corss_np=np.cross(vec1,vec2)

print(f"dot product:{dot_np}")
print(f"cross product: {corss_np}") ## |a|*|b|*sin()
```
