```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

factorial(5)
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
## transpose of matrix
def transpose(matrix):
    n= len(matrix)
    m= len(matrix[0])
    transposed = [[0 for j in range(n)] for i in range(m)]
    for i in range(n):
        for j in range(m):
            transposed[j][i] = matrix[i][j]
    return transposed

transpose(matrix)

## transpose of a matrix
def transpose_1(matrix):
    return [[row[i] for row in matrix] for i in range(len(matrix[0]))]

transpose_1(matrix)
```

```python
# Load the configuration files
extensions_file_path= "config/acr.txt"
lib_extensions= "config/extensions.txt"
include_extensions_path= "config/include_extensions.txt"
def read_config_files(filepath):
    known_acr= set()
    with open(filepath) as fp:
        obj= fp.readlines()
        for val in obj:
            val = val.strip()
            if val == '':
                continue
            known_acr.add(val)
        return known_acr

known_acr=read_config_files(extensions_file_path)
print(known_acr)
lib_extensions=read_config_files(lib_extensions)
print(lib_extensions)
include_ext= read_config_files(include_extensions_path)
print(include_ext)
```

```python
def load_configurations(file_path):
    config = {}
    current_section = None
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            if not line or line.startswith('#'):  # Skip empty lines and comments
                continue
            if line.startswith('[') and line.endswith(']'):
                current_section = line[1:-1].lower()  # Extract section name
                config[current_section] = set()
            else:
                config.setdefault(current_section, set()).add(line.lower())
    return config

# Load configurations from file
config_data = load_configurations('config/actual.txt')

# Assign to variables (for backward compatibility with existing code)
lib_extensions = config_data.get('lib', set())
include_extensions = config_data.get('include', set())
docs_extensions = config_data.get('docs', set())
help_extensions = config_data.get('help', set())
config_extensions = config_data.get('config', set())


print(lib_extensions)
print(include_extensions)
print(docs_extensions)
print(help_extensions)
print(config_extensions)
```

```python
import pefile

def normalize_string(s):
    """Convert to lowercase and remove all whitespace."""
    return s.lower().replace(" ", "")

def get_assembly_company_name(dll_path):
    try:
        pe = pefile.PE(dll_path)
        if not hasattr(pe, 'FileInfo'):
            return None
        
        for entry in pe.FileInfo:
            for child in entry:
                if hasattr(child, 'StringTable'):
                    for st in child.StringTable:
                        if b'CompanyName' in st.entries:
                            return st.entries[b'CompanyName'].decode('utf-8', errors='ignore')
        return None
    except Exception as e:
        print(f"Error reading DLL: {str(e)}")
        return None

def compare_company_name(dll_path, txt_path):
    # Extract CompanyName from DLL
    dll_company_name = get_assembly_company_name(dll_path)
    if not dll_company_name:
        return "CompanyName not found in DLL"
    
    # Read expected name from text file
    try:
        with open(txt_path, 'r') as f:
            expected_name = f.read().strip()  # Read entire file content
    except FileNotFoundError:
        return "Config file not found"
    except Exception as e:
        return f"Error reading config: {str(e)}"
    
    # Normalize both names
    normalized_dll = normalize_string(dll_company_name)
    normalized_txt = normalize_string(expected_name)
    
    # Compare
    if normalized_dll == normalized_txt:
        return "Match"
    else:
        return f"Mismatch: DLL='{normalized_dll}' vs TXT='{normalized_txt}'"

if __name__ == "__main__":
    dll_path = input("Enter DLL path: ")
    txt_path = "expected_name.txt"  # Update this path if needed
    result = compare_company_name(dll_path, txt_path)
    print(result)
```
