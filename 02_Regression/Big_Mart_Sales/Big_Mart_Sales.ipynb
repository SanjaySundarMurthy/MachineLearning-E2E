```python
X= df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
num_feat= X.select_dtypes(exclude="object").columns
cat_feat=X.select_dtypes(include="object").columns
print(num_feat, len(num_feat))
print(cat_feat, len(cat_feat))
cat_feat_nolbl= cat_feat.drop(["Item_Identifier","Outlet_Identifier"])
print(cat_feat_nolbl, len(cat_feat_nolbl))
```

```python
label_feat= ["Item_Identifier","Outlet_Identifier"]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
X["Item_Identifier"]=LabelEncoder().fit_transform(X["Item_Identifier"])
X["Outlet_Identifier"]=LabelEncoder().fit_transform(X["Outlet_Identifier"])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
cv = ColumnTransformer([
    ("onehotencoder",OneHotEncoder(),cat_feat_nolbl),
    ("standardScaler",StandardScaler(),num_feat)
])
cv
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=7)
print(X_train.shape, y_train.shape, X_test.shape,y_test.shape)
```

```python
X_train= cv.fit_transform(X_train)
X_test= cv.transform(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
def evaluate_metrics(y_test,y_pred):
  r2= r2_score(y_test,y_pred)
  mae= mean_absolute_error(y_test,y_pred)
  return r2, mae
```

```python
from sklearn.linear_model import LinearRegression,ElasticNet,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
models= {
    "LinearRegression": LinearRegression(),
    "LinearRegression": LinearRegression(),
    "Ridge":Ridge(),
    "RandomForestRegressor": RandomForestRegressor(),
    "AdaBoostRegressor": AdaBoostRegressor(),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "SVR": SVR(),
    "Lasso": Lasso(),
    "XGB": XGBRegressor()
    # "DecisionTreeRegressor": DecisionTreeRegressor()
}
def evaluate_models(models,X_train,y_train,X_test,y_test):
  results={}
  for i in range(len(list(models))):
    name= list(models.keys())[i]
    model= list(models.values())[i]

    ## Fit the data
    model.fit(X_train,y_train)
    #predict
    y_pred= model.predict(X_test)

    ## Metrics
    r2, mae= evaluate_metrics(y_test,y_pred)
    print(f"name: {name}")
    print("*"*50)
    print(f"R2 Score: {r2}")
    print(f"MAE : {mae}")
    print("*"*50)
    results[name]= {'test': {'mae': mae,'r2':r2}}
  return results

```

```python
def plot_results(results):
  model_names = list(results.keys())
  # train_r2_scores = [results[name]['train']['R2'] for name in model_names]
  test_r2_scores = [results[name]['test']['r2'] for name in model_names]

  plt.figure(figsize=(10, 5))
  # sns.barplot(x=model_names, y=train_r2_scores, color='blue', alpha=0.6, label='Train R2')
  sns.barplot(x=model_names, y=test_r2_scores, color='red', alpha=0.6, label='Test R2')
  plt.xticks(rotation=45)
  plt.title("Model R2 Scores for Train and Test Sets")
  plt.ylabel("R2 Score")
  plt.legend()
  plt.show()
```

```python
params={
    "RandomForestRegressor": {
        "n_estimators": [50, 100, 150],
        "max_depth": [10, 20, 30],
        "min_samples_split": [2, 5, 10],
        "min_samples_leaf": [1, 2, 4]
    },
    "GradientBoostingRegressor": {
        "n_estimators": [50, 100, 150],
        "learning_rate": [0.01, 0.1, 0.2],
        "max_depth": [3, 5, 7]
    },
    "AdaBoostRegressor": {
        "n_estimators": [50, 100, 150],
        "learning_rate": [0.01, 0.1, 1]
    },
    "SVR": {
        "C": [0.1, 1, 10],
        "epsilon": [0.01, 0.1, 1],
        "kernel": ["linear", "poly", "rbf"]
    },
    "KNeighborsRegressor": {
        "n_neighbors": [3, 5, 7, 9],
        "weights": ["uniform", "distance"],
        "p": [1, 2]
    },
    "LinearRegression": {},
    # "Lasso": {
    #     "alpha": [0.01, 0.1, 1, 10],
    #     "max_iter": [1000, 2000, 5000]
    # },
    "Ridge": {
        "alpha": [0.01, 0.1, 1, 10],
        "max_iter": [1000, 2000, 5000]
    }
}
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV , StratifiedKFold
def hyperparams_tune(models,X_train,y_train,params):
  best_params={}
  for i in range(len(list(models))):
    name= list(models.keys())[i]
    model= list(models.values())[i]
    cv=5
    if name in params.keys():
      param_val= params[name]
      gs= RandomizedSearchCV(estimator=model, param_distributions=param_val, scoring="r2",cv=5,verbose=2,n_jobs=-1)
      gs.fit(X_train,y_train)
      print(f"Best params:{gs.best_params_}")
      print(f"Best score: {gs.best_score_}")
      models[name].set_params(**gs.best_params_)
      best_params[name]= gs.best_params_
      # print(best_params)
    else:
      print(f"No params found for this model: {name}")
  # print(models)
  return models


models=hyperparams_tune(models,X_train,y_train,params)
print(models)
results=evaluate_models(models,X_train,y_train,X_test,y_test)
print(results)
plot_results(results)

```
