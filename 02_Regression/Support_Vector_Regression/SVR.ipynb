```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_classes=2, n_clusters_per_class=1, n_features=2, n_informative=2, n_redundant=0,random_state=7)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=7)
```

```python
from sklearn.svm import SVR
model=SVR()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

print(f"R2: score: {r2_score(y_test,y_pred)}")
print(f"MAE: score: {mean_absolute_error(y_test,y_pred)}")
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.30,random_state=7)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

le1=LabelEncoder()
cat= ["sex","time","smoker"]

for i in range(len(cat)):
  X_train[cat[i]]=le1.fit_transform(X_train[cat[i]])
  X_test[cat[i]]=le1.transform(X_test[cat[i]])
```

```python
X_train[:10]
```

```python
X_test[:10]
```

```python
cv= ColumnTransformer(transformers=[("onehot", OneHotEncoder(drop="first"),[3])], remainder="passthrough")
X_train=cv.fit_transform(X_train)
```

```python
X_test=cv.transform(X_test)
```

```python
from sklearn.svm import SVR
model_sv=SVR(kernel="linear")
model_sv.fit(X_train,y_train)

y_pred_sv=model_sv.predict(X_test)
from sklearn.metrics import r2_score, mean_absolute_error

print(f"R2: score: {r2_score(y_test,y_pred_sv)}")
print(f"MAE: score: {mean_absolute_error(y_test,y_pred_sv)}")
```

## Hyper parametr tuning

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)

# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction=grid.predict(X_test)

from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,grid_prediction))
print(mean_absolute_error(y_test,grid_prediction))
```

```python
from sklearn.svm import SVR
model_sv=SVR(C=1000, gamma=0.0001, kernel='rbf')
model_sv.fit(X_train,y_train)

y_pred_sv=model_sv.predict(X_test)
from sklearn.metrics import r2_score, mean_absolute_error

print(f"R2: score: {r2_score(y_test,y_pred_sv)}")
print(f"MAE: score: {mean_absolute_error(y_test,y_pred_sv)}")
```
