```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sklearn
```

```python
from sklearn.datasets import load_iris

data = load_iris()
data
```

```python
df=pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
y=data.target
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(df,y,test_size=0.25, random_state=7)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_cls= DecisionTreeClassifier()
model_cls.fit(X_train,y_train)

y_pred_cls= model_cls.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

print(f"accuarcy score: {accuracy_score(y_test,y_pred_cls)}")
print(f"classification Report: {classification_report(y_test,y_pred_cls)}")

```

```python
from sklearn import tree
plt.figure(figsize=(10,7))
tree.plot_tree(model_cls, filled=True)
plt.show()
```

```python
input_data=(5.9, 3. , 5.1, 1.8)
data= np.asarray(input_data).reshape(1,-1)
prediction= model_cls.predict(data)

if prediction[0]==0:
  print("setosa")
elif prediction[0]==1:
  print("versicolor")
else:
  print("virginica")
```

## Regression

```python
from sklearn.datasets import load_diabetes

data=load_diabetes()

df=pd.DataFrame(data.data, columns=data.feature_names)

df.head()
```

```python
y=data.target
y[:10]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(df,y,test_size=0.25, random_state=7)
```

```python
from sklearn.tree import DecisionTreeRegressor

model_reg= DecisionTreeRegressor()
model_reg.fit(X_train,y_train)

y_pred_reg= model_reg.predict(X_test)

from sklearn.metrics import mean_absolute_error, r2_score

print(f"R2 score: {r2_score(y_test,y_pred_reg)}")
print(f"MAE : {mean_absolute_error(y_test,y_pred_reg)}")
```
