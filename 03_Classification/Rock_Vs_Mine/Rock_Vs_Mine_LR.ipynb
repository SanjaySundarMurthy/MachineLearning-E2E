```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.10,stratify=y,random_state=7)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
model2=LogisticRegression()
model2.fit(X_train,y_train)

y_pred_2=model2.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

print(f"Accuarcay: {accuracy_score(y_test,y_pred_2)}")
print(f"Report: {classification_report(y_test,y_pred_2)}")
```

```python
params= {
        "C": [0.01, 0.1, 1, 10],
        "solver": ["liblinear", "saga"],
        "max_iter": [100, 200, 500]
    }

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

gs=GridSearchCV(estimator=model2, param_grid=params, cv=5, scoring="accuracy",refit=True,verbose=2, n_jobs=-1)
gs.fit(X_train,y_train)

print(f"Best score: {gs.best_score_}")
print(f"Best params: {gs.best_params_}")
model3=LogisticRegression()
model3.set_params(**gs.best_params_)
model3.fit(X_train,y_train)
y_pred_2=model3.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report

print(f"Accuarcay: {accuracy_score(y_test,y_pred_2)}")
print(f"Report: {classification_report(y_test,y_pred_2)}")
```

```python
rnd= RandomizedSearchCV(estimator=model2,param_distributions=params,scoring="roc_auc",refit=True,verbose=2, n_jobs=-1)
rnd.fit(X_train,y_train)
print(f"Best score: {rnd.best_score_}")
print(f"Best params: {rnd.best_params_}")

model3.set_params(**rnd.best_params_)
model3.fit(X_train,y_train)
y_pred_2=model3.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report

print(f"Accuarcay: {accuracy_score(y_test,y_pred_2)}")
print(f"Report: {classification_report(y_test,y_pred_2)}")
```
