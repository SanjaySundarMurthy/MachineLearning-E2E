## Converting RGB images to GrayScale images

```python
import cv2

img= cv2.imread('/content/dog_image.jpg')
print(type(img))
print(img.shape)
```

```python
grayscale_img= cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
print(type(grayscale_img))
print(grayscale_img.shape)
```

```python
plt.imshow(grayscale_img)
plt.show()
```

```python
## cv2.imshow() --> this will not be allowed in google collab but it works in alternate

from google.colab.patches import cv2_imshow
cv2_imshow(grayscale_img)
```

```python
## Saving the grayscale image
cv2.imwrite('dog_gray_scale.jpg', grayscale_img)
```
