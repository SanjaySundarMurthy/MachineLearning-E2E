```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_breast_cancer
```

```python
data= load_breast_cancer()
data
```

```python
df= pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
target= data.target
target
```

```python
X= df
y=target
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.20, random_state=7)
print(X_train.shape, X_test.shape, y_train.shape,y_test.shape)
```

```python
## Standardize the data
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
## building a Neural network
import tensorflow as tf
tf.random.set_seed(3) ## several weights will be assigned for each time we run so to have the accuracy same we are settting the random seed
from tensorflow import keras
```

```python
## Setting up the layers
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(2, activation="sigmoid")
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
## Sparse categorical cross entropy ---> for label encoder , or with o's and 1's
## Categorical cross entropy --> onehot encoder
```

```python
## training the model
history= model.fit(X_train,y_train,validation_split=0.1,epochs=10)
```

```python
model.summary()
```

```python
history.history
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','val'])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','val'])
```

```python
## Accuarcy of the value on test data

loss, accuracy= model.evaluate(X_test,y_test)

print(accuracy)
```
